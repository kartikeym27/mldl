```python
import pandas as pd
import time
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# %matplotlib inline
import plotly.express as px
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
df = pd.read_csv('/content/water_potability.csv')
X = df.drop('Potability', axis=1)
y = df['Potability']
X.shape, y.shape




# Training and Testing Data




from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)




# Centralized Decision Tree Classifier




from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train,y_train)
pred_dt = model_dt.predict(X_test)
dt = accuracy_score(y_test, pred_dt)
print(classification_report(y_test,pred_dt))




# Random Forest Classifier




from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=300,min_samples_leaf=0.16)
model_rf.fit(X_train, y_train)
pred_rf = model_rf.predict(X_test)
rf = accuracy_score(y_test, pred_rf)
print(classification_report(y_test,pred_rf))




# Distributed Decision Tree Classifier
x1 = X_train[["ph","Solids","Sulfate","Organic_carbon","Turbidity"]]
x2 = X_train[["ph","Hardness","Solids","Chloramines","Sulfate"]]
x3 = X_train[["Conductivity","Organic_carbon","Trihalomethanes","Turbidity"]]
X_test_1 = X_test[["ph","Solids","Sulfate","Organic_carbon","Turbidity"]]
X_test_2 = X_test[["ph","Hardness","Solids","Chloramines","Sulfate"]]
X_test_3 = X_test[["Conductivity","Organic_carbon","Trihalomethanes","Turbidity"]]


model_dt1 = DecisionTreeClassifier()
model_dt1.fit(x1,y_train)
model_dt2 = DecisionTreeClassifier()
model_dt2.fit(x2,y_train)
model_dt3 = DecisionTreeClassifier()
model_dt3.fit(x3,y_train)
pred_dt1 = model_dt1.predict(X_test_1)
pred_dt2 = model_dt2.predict(X_test_2)
pred_dt3 = model_dt3.predict(X_test_3)
dt1 = accuracy_score(y_test, pred_dt1)
dt2 = accuracy_score(y_test, pred_dt2)
dt3 = accuracy_score(y_test, pred_dt3)
predictions = np.stack([pred_dt1, pred_dt2, pred_dt3], axis=1)
from scipy.stats import mode
final_predictions,_ = mode(predictions, axis=1)
final_predictions = final_predictions.ravel()
fdt = accuracy_score(y_test, final_predictions)
print(classification_report(y_test,final_predictions))


```
