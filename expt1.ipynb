```python
from numpy import random
import matplotlib.pyplot as plt


import seaborn as sns
sns.distplot(random.chisquare(df=1, size=1000), hist=False)
plt.show()
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, chisquare


# Parameters for the Poisson distribution
lambda_param = 5  # Mean of the distribution
size = 1000       # Number of samples


# Generate data from a Poisson distribution
data = np.random.poisson(lambda_param, size)


# Calculate the observed frequencies
bins = np.arange(0, np.max(data) + 2) - 0.5
observed_freq, _ = np.histogram(data, bins=bins)


# Calculate the expected frequencies
x = np.arange(0, np.max(data) + 1)
expected_freq = poisson.pmf(x, lambda_param) * size


# Ensure both arrays have the same length by aligning with the same bins
if len(expected_freq) < len(observed_freq):
    expected_freq = np.append(expected_freq, [0] * (len(observed_freq) - len(expected_freq)))


# Check that the sums of observed and expected frequencies are equal
observed_sum = np.sum(observed_freq)
expected_sum = np.sum(expected_freq)


if abs(observed_sum - expected_sum) > 1e-8:
    # Adjust expected frequencies to match the observed sum
    expected_freq *= observed_sum / expected_sum


# Perform the Chi-Square goodness-of-fit test
chi_statistic, p_value = chisquare(observed_freq, expected_freq)
print(f"Chi-Square Statistic: {chi_statistic}")

print(f"P-value: {p_value}")


# Plot the histogram of the data
plt.hist(data, bins=bins, density=True, alpha=0.7, color='orange', edgecolor='black')


# Plot the theoretical Poisson distribution PMF
plt.plot(x, poisson.pmf(x, lambda_param), 'r-', lw=2, label='Theoretical Poisson PMF')


plt.title('Poisson Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
from scipy.stats import expon, kstest




scale = 2.0
size = 100












data = np.random.exponential(scale, size)




plt.hist(data, bins=50, density=True, alpha=0.7)




x = np.linspace(0, np.max(data), 100)
pdf = expon.pdf(x, scale=scale)
plt.plot(x, pdf, 'r-', lw=2)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()




ks_statistic, p_value = kstest(data, 'expon', args=(0, scale))
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, chisquare


# Parameters for the Binomial distribution
n_trials = 10    # Number of trials
p_success = 0.5  # Probability of success
size = 1000      # Number of samples


# Generate data from a Binomial distribution
data = np.random.binomial(n_trials, p_success, size)


# Calculate the observed frequencies
bins = np.arange(-0.5, n_trials + 1.5, 1)  # Bins for each possible outcome
observed_freq, _ = np.histogram(data, bins=bins)


# Calculate the expected frequencies
x = np.arange(0, n_trials + 1)
expected_freq = binom.pmf(x, n_trials, p_success) * size


# Perform the Chi-Square goodness-of-fit test
chi_statistic, p_value = chisquare(observed_freq, expected_freq)
print(f"Chi-Square Statistic: {chi_statistic}")
print(f"P-value: {p_value}")


# Plot the histogram of the data
plt.hist(data, bins=bins, density=True, alpha=0.7, color='lightblue', edgecolor='black')


# Plot the theoretical Binomial distribution PMF
plt.plot(x, binom.pmf(x, n_trials, p_success), 'r-', lw=2, label='Theoretical Binomial PMF')


plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
